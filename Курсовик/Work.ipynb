## Описание датасет
Эксперименты были проведены с группой из 30 добровольцев в возрасте от 19 до 48 лет. Каждый человек выполнял шесть видов деятельности (ХОДЬБА, ПОДЪЕМ ПО ЛЕСТНИЦЕ, СПУСК ПО ЛЕСТНИЦЕ, СИДЕНИЕ, СТОЯНИЕ, ЛЕЖАНИЕ), нося смартфон (Samsung Galaxy S II) на поясе. С помощью встроенных в него акселерометра и гироскопа мы фиксировали трехосевое линейное ускорение и трехосевую угловую скорость с постоянной частотой 50 Гц. Эксперименты были записаны на видео для ручного маркирования данных. Полученный набор данных был случайным образом разделен на две части: 70% добровольцев было выделено для создания обучающих данных, а 30% — для тестовых данных.

Сигналы сенсоров (акселерометра и гироскопа) были предварительно обработаны с применением фильтров шума, а затем сегментированы в скользящие окна фиксированной ширины 2,56 секунды с 50% перекрытием (128 измерений на окно). Сигнал ускорения с сенсора, содержащий компоненты гравитации и движения тела, был разделен с помощью низкочастотного фильтра Баттерворта на ускорение тела и гравитацию. Предполагается, что гравитационная сила содержит только низкочастотные компоненты, поэтому использовался фильтр с частотой среза 0,3 Гц. Из каждого окна был получен вектор признаков путем вычисления переменных во временной и частотной областях.

## Загрузка данных

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import requests
from io import StringIO
from sklearn.metrics import confusion_matrix, precision_score, recall_score, classification_report
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

import seaborn as sns
import zipfile
import io
```

```python
# URL для скачивания данных
url = "https://archive.ics.uci.edu/static/public/240/human+activity+recognition+using+smartphones.zip"

# Скачивание содержимого файла
response = requests.get(url)
response.raise_for_status()  # Проверка на ошибки
```

```python
with io.BytesIO(response.content) as z_file:
    with zipfile.ZipFile(z_file, 'r') as outer_zip_ref:
        # Получите список файлов внутри внешнего архива
        file_list = outer_zip_ref.namelist()

        # Вывод списка файлов внешнего архива
        print("Files in the outer ZIP archive:")
        for file_name in file_list:
            print(file_name)

        # Проверим, содержится ли вложенный ZIP-файл
        if 'UCI HAR Dataset.zip' in file_list:
            # Открываем вложенный ZIP-файл
            with outer_zip_ref.open('UCI HAR Dataset.zip') as inner_zip_bytes:
                with zipfile.ZipFile(inner_zip_bytes, 'r') as inner_zip_ref:


                    # Предположим, что внутри ZIP лежит CSV-файл
                    with inner_zip_ref.open('UCI HAR Dataset/train/X_train.txt') as f:
                        X_train = pd.read_csv(f, sep='\\s+', header=None)
                
                    with inner_zip_ref.open('UCI HAR Dataset/train/y_train.txt') as f:
                        y_train = pd.read_csv(f, sep='\\s+', header=None)
                
                    with inner_zip_ref.open('UCI HAR Dataset/test/y_test.txt') as f:
                        y_test = pd.read_csv(f, sep='\\s+', header=None)
                
                    with inner_zip_ref.open('UCI HAR Dataset/test/X_test.txt') as f:
                        X_test = pd.read_csv(f, sep='\\s+', header=None)
                
                    with inner_zip_ref.open('UCI HAR Dataset/features.txt') as f:
                        features_name = pd.read_csv(f, sep='\\s+', header=None)
```

```python
 
X = pd.concat([X_train, X_test], axis = 0)
y = pd.concat([y_train, y_test], axis = 0)
y.columns = ['target']
X.columns = features_name[1].values

df = pd.concat([y, X], axis = 1)
df.to_csv('dataset.csv', index = False)

    
```

### 3. Выбор задачи и алгоритма

Для выбора задачи и алгоритма, идентифицируйте, какая задача соответствует вашим данным. В данном случае, если target является категориальным (классы активности), то стоит выбрать задачу классификации.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Разделение данных на обучающую и тестовую выборки
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Инициализация и обучение модели Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Предсказание и оценка
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Точность модели: {accuracy:.2f}")
```

```python
# Метрика производительности
print("Classification Report:\n", classification_report(y_test, y_pred))

# Вычисление и вывод confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Визуализация confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Вычисление и вывод precision и recall
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print(f"Precision (Точность): {precision:.2f}")
print(f"Recall (Полнота): {recall:.2f}")
```

```python
# Инициализация и обучение SVM
svm_model = SVC(kernel='rbf', C=1, gamma='scale', random_state=42)
svm_model.fit(X_train, y_train)

# Предсказание классов на тестовой выборке
y_pred = svm_model.predict(X_test)

```

```python
# Метрика производительности
print("Classification Report:\n", classification_report(y_test, y_pred))

# Вычисление и вывод confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Визуализация confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Вычисление и вывод precision и recall
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print(f"Precision (Точность): {precision:.2f}")
print(f"Recall (Полнота): {recall:.2f}")
```

## 4. Кластеризация/Понижение размерности

Используйте алгоритмы понижения размерности, такие как PCA (анализ главных компонент), и методы кластеризации, такие как K-Means, для изучения данных.

```python
# Диапазон возможных кластеров
range_n_clusters = list(range(1, 11))
inertia = []

# Вычисляем инерцию для каждого k
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Построение графика метода локтя
plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, inertia, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')
plt.title('Метод локтя для K-Means')
plt.show()
```

```python
range_n_clusters = list(range(2, 11))

silhouette_avg = []
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg.append(silhouette_score(X, cluster_labels))

# Построение графика силуэтного анализа
plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, silhouette_avg, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Средняя силуэтная оценка')
plt.title('Силуэтный метод для K-Means')
plt.show()
```

```python
n_clusters = 6

# Кластеризация с использованием K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
# Понижение размерности с использованием PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Визуализация категорий
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Spectral', s=2)
plt.title('Распределения данных после PCA и по текущим классам')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
# Визуализация кластеров
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='Spectral', s=2)
plt.title('Распределения данных после PCA и K-Means кластеризации')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
